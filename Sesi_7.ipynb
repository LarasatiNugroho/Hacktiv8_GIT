```python
import numpy as np
import pandas as pd
```

```python
df_can = pd.read_excel('https://raw.githubusercontent.com/ardhiraka/PFDS_sources/master/Canada.xlsx'
                      , sheet_name='Canada by Citizenship'
                      , skiprows=range(20)
                      , skipfooter=2)
print("Data read into a pandas dataframe")
```

```python
df_can.head()
```

```python
df_can.drop(
    ["AREA","REG","DEV","Type","Coverage"]
    ,axis = 1, inplace = True
    )

df_can.head(3)
```

```python
# ganti nama column
df_can.rename(
    columns={
        "OdName" : "Country",
        "AreaName" : "Continent",
        "RegName" : "Region",
    }, inplace=True
    )

df_can.head(3)
```

```python
# axis 1 = mendatar
df_can["Total"] = df_can.sum(axis=1)

df_can.head(2)
```

```python
df_can.set_index("Country", inplace=True)
```

```python
df_can.head(2)
```

```python
df_can.loc["Indonesia"]
```

```python
df_can.columns = list(map(str,df_can.columns)) 
#buat list tahun 1980-2013 diubah jadi str, 
#trus hasilnya dimasukin ke variable df_can.columns.
#Pakai funtion map, yang tujuannya mirip dengan loop
```

```python
years = list(map(str,range(1980, 2014))) #max 2014,biar 2013 nya masuk

years
```

```python
df_can[
    df_can["Continent"] == "Asia"
]
```

```python
# namanya query untuk explore
df_can[
    df_can["Region"] == "South-Eastern Asia"
]
```

## Pie Chart

```python
df_continents = df_can.groupby("Continent",axis=0).sum()
df_continents
```

```python
explode_list = [0.1,0,0,0,0.1,0.1]

df_continents["Total"].plot(
    kind="pie", 
    figsize = (15,6),
    autopct = "%1.1f%%", #1 angka dibelakang koma
    startangle = 90, # means urutan dari df_can
    shadow = True, #tambah shadow
    labels = None, #nggak ada label nama
    pctdistance = 1.12,
    explode = explode_list #buat kayak potongan pizza
)

plt.title("Immigrants to Canada by Continent", y=1.12) #nambah 1.12 dr awal ke atas
plt.axis("Equal") #biar lokasinya benaran ditengah

plt.legend(labels =df_continents.index, loc="upper left" )

plt.show()
```

## Sub Plot

```python
df_tot = pd.DataFrame(df_can[years].sum(axis=0)) #axis=0 vertical

df_tot.head()
```

```python
#ubah index dr str ke int

df_tot.index = map(int,df_tot.index)
df_tot.reset_index(inplace=True)
df_tot.head()
```

```python
df_tot.columns = ["Years", "Total"]
df_tot.head()
```

```python
df_tot.plot(
    kind="scatter",
    x="Years",
    y="Total"
    )

plt.title("Years vs Total Immigrants")
plt.xlabel("Years")
plt.ylabel("Total Immigrants around the world")

plt.show()
```
