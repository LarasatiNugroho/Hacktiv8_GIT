```python
from wordcloud import WordCloud, STOPWORDS # import
```

```python
#download file; https://raw.githubusercontent.com/ardhiraka/PFDS_sources/master/alice_novel.txt
```

```python
#hilangin stop words: cth is, the, 
```

```python
alice_novel = open("dataset/alice_novel.txt","r").read() #"r" artinya read
```

```python
stopwords = set(STOPWORDS)
```

```python
alice_wc = WordCloud(
    background_color = "white", #background colors
    stopwords = stopwords, #kata yg perlu dihilangin
    max_words = 2000 #banyaknya kata utk dianggap maximum, semakin banyak semakin berat
)

alice_wc.generate(alice_novel)
```

```python
plt.imshow(alice_wc, interpolation="bilinear") #interpolation yaitu paremter untuk treat gambarr supaya lebih halus

plt.axis("off") #untuk matiin grid
plt.show()
```

```python
fig = plt.figure()
fig.set_figwidth(14)
fig.set_figheight(118)

plt.imshow(alice_wc, interpolation="bilinear") #interpolation yaitu paremter untuk treat gambarr supaya lebih halus

plt.axis("off") #untuk matiin grid
plt.show()
```

```python
#cara nambahin stopwords 
stopwords.add("said")

alice_wc.generate(alice_novel)
plt.imshow(alice_wc, interpolation="bilinear") #interpolation yaitu paremter untuk treat gambarr supaya lebih halus

plt.axis("off") #untuk matiin grid
plt.show()
```

## Regression Plot

```python
!pip install seaborn
```

```python
import seaborn as sns
```

```python
df_tot = pd.DataFrame(df_can[years].sum(axis=0))
```

```python
df_tot.index = map(int,df_tot.index)
```

```python
df_tot.reset_index(inplace=True)
df_tot.columns = ["year","total"]
```

```python
df_tot.head()
```

```python
# dokumentasi seaborn: library https://seaborn.pydata.org/api.html
```

```python
ax = sns.regplot(x="year",y="total",data=df_tot)
```
